### **Problem Statement**

The objective is to load, inspect, and visualize the Iris dataset using different graphs such as scatter plots, histograms, and box plots to understand relationships, distributions, and outliers in the data.

# **DATA LOADING**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the built-in iris dataset
df = sns.load_dataset('iris')

print(df.head())
```

# **DATA CLEANING AND VISULAIZATION:**

```python
df.isnull().sum()
```

```python
sns.scatterplot(x="sepal_length",y="petal_length",hue="species",data=df)
plt.title("Scatter Plot of Sepal Length vs Petal Length")
plt.show()
```

```python
sns.scatterplot(x="sepal_width",y="petal_width",hue="species",data=df)
plt.title("Scatter Plot of Sepal width vs Petal width")
plt.show()
```

```python
plt.hist(df['sepal_length'], bins=10)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df['sepal_width'], bins=10)
plt.title("Histogram of Sepal width")
plt.xlabel("Sepal width")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df['petal_length'],bins=10,color="black")
plt.title("Box Plot of petal Length")
plt.ylabel("petal Length")
plt.show()
```

```python
plt.hist(df['petal_width'],bins=10)
plt.title("Box Plot of petal width")
plt.ylabel("petal width")
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```

```python
sns.boxplot(x='species', y='petal_length', data=df)
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```
